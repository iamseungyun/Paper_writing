```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
from matplotlib import pyplot as plt
import pickle
import import_ipynb
import os
```

```python
path = os.getcwd() + '/result/income_replacement/70/'
```

```python
p_val = pd.read_pickle(path +'p_val.pkl')
w_val = pd.read_pickle(path +'w_val.pkl')
beq = pd.read_pickle(path +'bequest.pkl')
inc = pd.read_pickle(path +'income.pkl')
dep = pd.read_pickle(path +'depletion.pkl')
```

```python
bequest_20_m = []

for col in beq.columns[:9] :
    bequest_20_m.append(beq[col]['minimum'].mean().mean())

bequest_25_m = []

for col in beq.columns[9:18] :
    bequest_25_m.append(beq[col]['minimum'].mean().mean())
    
bequest_30_m = []

for col in beq.columns[18:27] :
    bequest_30_m.append(beq[col]['minimum'].mean().mean())
```

```python
income_20_m = []

for col in inc.columns[:9] :
    income_20_m.append(inc[col]['minimum'].mean().mean()/1000)

income_25_m = []

for col in inc.columns[9:18] :
    income_25_m.append(inc[col]['minimum'].mean().mean()/1000)

income_30_m = []

for col in inc.columns[18:27] :
    income_30_m.append(inc[col]['minimum'].mean().mean()/1000)
```

```python
depletion_20_m = []

for col in dep.columns[:9] :
    depletion_20_m.append(dep[col]['minimum'].mean().mean())

depletion_25_m = []

for col in dep.columns[9:18] :
    depletion_25_m.append(dep[col]['minimum'].mean().mean())

depletion_30_m = []

for col in dep.columns[18:27] :
    depletion_30_m.append(dep[col]['minimum'].mean().mean())
```

```python
bequest_20_p = []

for col in beq.columns[:9] :
    bequest_20_p.append(beq[col]['proper'].mean().mean())

bequest_25_p = []

for col in beq.columns[9:18] :
    bequest_25_p.append(beq[col]['proper'].mean().mean())

bequest_30_p = []

for col in beq.columns[18:27] :
    bequest_30_p.append(beq[col]['proper'].mean().mean())
```

```python
income_20_p = []

for col in inc.columns[:9] :
    income_20_p.append(inc[col]['proper'].mean().mean()/1000)

income_25_p = []

for col in inc.columns[9:18] :
    income_25_p.append(inc[col]['proper'].mean().mean()/1000)

income_30_p = []

for col in inc.columns[18:27] :
    income_30_p.append(inc[col]['proper'].mean().mean()/1000)
```

```python
depletion_20_p = []

for col in dep.columns[:9] :
    depletion_20_p.append(dep[col]['proper'].mean().mean())

depletion_25_p = []

for col in dep.columns[9:18] :
    depletion_25_p.append(dep[col]['proper'].mean().mean())

depletion_30_p = []

for col in dep.columns[18:27] :
    depletion_30_p.append(dep[col]['proper'].mean().mean())
```

```python
labels = ['p10','p20','p30','p40','p50','p60','p70','p80', 'p90']
x = np.arange(len(labels))
```

```python
bar_width = 0.3
    
plt.figure(figsize=(12,6))
plt.bar(x-0.2, bequest_20_m, bar_width, alpha=0.4, color = 'blue', label = 'Minimum spending')
plt.bar(x+bar_width-0.05, bequest_20_p, bar_width, alpha=0.4, color='green', label = 'Proper spending') 
plt.xticks(x, labels)
plt.ylabel('Bequest(percentage)')
plt.ylim((0,100))
plt.legend(bbox_to_anchor = (1.05,1.0), loc='upper left')
plt.title('Bequest after 20 years')
plt.show()

plt.figure(figsize=(12,6))
plt.bar(x-0.2, income_20_m, bar_width, alpha=0.4, color = 'blue', label = 'Minimum spending')
plt.bar(x+bar_width-0.05, income_20_p, bar_width, alpha=0.4, color='green', label = 'Proper spending') 
plt.xticks(x, labels)
plt.ylabel('Total income(thousands)')
plt.ylim((0,15))
plt.legend(bbox_to_anchor = (1.05,1.0), loc='upper left')
plt.title('Total income after 20 years')
plt.show()

plt.figure(figsize=(12,6))
plt.bar(x-0.2, depletion_20_m, bar_width, alpha=0.4, color = 'blue', label = 'Minimum spending')
plt.bar(x+bar_width-0.05, depletion_20_p, bar_width, alpha=0.4, color='green', label = 'Proper spending') 
plt.xticks(x, labels)
plt.ylabel('Portfolio depletion probability')
plt.ylim((0,1))
plt.legend(bbox_to_anchor = (1.05,1.0), loc='upper left')
plt.title('Portfolio depletion probability after 20 years')
plt.show()
```
