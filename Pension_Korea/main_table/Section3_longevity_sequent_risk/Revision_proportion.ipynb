```python
import matplotlib.pyplot as plt
import numpy as np

# Data
quarters = ["2021 1Q", "2021 2Q", "2021 3Q", "2021 4Q", "2022 1Q", "2022 2Q", "2022 3Q", "2022 4Q"]
not_guranteed = [0.474942503, 0.520900196, 0.53800031, 0.534863109, 0.501482796, 0.465934572, 0.460244245, 0.432324121]
guranteed = [0.525057497, 0.479099804, 0.46199969, 0.465136891, 0.498517204, 0.534065428, 0.539755755, 0.567675879]
plt.figure(figsize=(12, 6))
# Create stacked bar chart
bar_width = 0.5
avg_data_1 = sum(not_guranteed) / len(not_guranteed)
plt.bar(quarters, not_guranteed, width = bar_width, label="(a) Non-principal guaranteed", color="#dedbd2")
plt.bar(quarters, guranteed, width = bar_width, bottom=not_guranteed, label="(b) Principal guaranteed", color = "#edafb8")
plt.axhline(y=avg_data_1, color="#4a5759", linestyle="--", label="Average of (a)")

plt.grid(axis='both', linestyle='--', linewidth=0.5, alpha=0.6)
# Labeling and showing
plt.xlabel("Time", fontsize=18)
plt.ylabel("Proportion", fontsize=18)
# plt.title("Proportion of IRP reserves of securities and financial investment companies", fontsize=20)
plt.ylim(0, 1.4)
plt.legend(fontsize=14)
plt.tight_layout()
plt.savefig("proportion_IRP.eps", format="eps")
plt.show()


```
