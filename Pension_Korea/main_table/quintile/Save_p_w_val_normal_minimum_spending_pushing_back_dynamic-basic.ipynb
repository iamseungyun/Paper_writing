```python
df =  pd.read_csv('./../data/data.csv')
portfolio_strategy = [static]
upper_years = [10,15,20, 25, 30]
periods = [30, 35, 40]
```

```python
init_p_vals = init_wealth_list
bear_year = np.load('./../result/year/bear_year.npy', allow_pickle=True).reshape(1)[0]
```

# normal

```python
normal_withdrawal_rate = np.array([[0.12954898],
       [0.12230136],
       [0.11505374]])
```

```python
functions = [dyna_withdraw]

for p, period in enumerate(periods) :
    for upper_year in upper_years: # 5가지
        for k in [0.5]:
            for i, withdraw_rate in enumerate(list(normal_withdrawal_rate[p])):
                print('year{}_k{}_rate{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, withdraw_rate, i+1, period))
                p_val = {}
                w_val = {}
                years = bear_year[upper_year][0]
                for j, weight in enumerate(portfolio_strategy) :
                    print('{} Strategy'.format(weight))

                    for function in functions :
                        for year in years :
                            print(year)
                            res = function(year, weight, k, upper_year, withdraw_rate, df, init_p_vals[i])
                            
                            p_val[year] = res[0]
                            w_val[year] = res[1]

                p_val = pd.DataFrame(p_val)
                w_val = pd.DataFrame(w_val)
                p_val.to_csv('./../result/late_retire/p_val/year{}_k{}_decile{}_period{}_const_normal_spending_dynamic_basic.csv'.format(upper_year, k, i+1, period))
                w_val.to_csv('./../result/late_retire/w_val/year{}_k{}_decile{}_period{}_const_normal_spending_dynamic_basic.csv'.format(upper_year, k, i+1, period))

```

```python
p_val = {}
w_val = {}
for p, period in enumerate(periods) :
    for upper_year in upper_years:
        for k in [0.5]:
            for i, withdraw_rate in enumerate(list(normal_withdrawal_rate[p])):
                p_val['year{}_k{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, i+1, period)] = {}
                w_val['year{}_k{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, i+1, period)] = {}
```

```python

for p, period in enumerate(periods) :
    for upper_year in upper_years:
        for k in [0.5]:
            for i, withdraw_rate in enumerate(list(normal_withdrawal_rate[p])):
                years = bear_year[upper_year][0]
                for year in years:
                    print('year{}_k{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, i+1, period))
                    print(year)
                    p_val['year{}_k{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, i+1, period)][year] = np.array(pd.read_csv('./../result/late_retire/p_val/year{}_k{}_decile{}_period{}_const_normal_spending_dynamic_basic.csv'.format(upper_year, k, i+1, period))[str(year)])
                    w_val['year{}_k{}_decile{}_period{}_dynamic_basic'.format(upper_year, k, i+1, period)][year] = np.array(pd.read_csv('./../result/late_retire/w_val/year{}_k{}_decile{}_period{}_const_normal_spending_dynamic_basic.csv'.format(upper_year, k, i+1, period))[str(year)])

np.save('./../result/late_retire/p_val/p_val_const_normal_spending_dynamic_basic.npy', p_val)
np.save('./../result/late_retire/w_val/w_val_const_normal_spending_dynamic_basic.npy', w_val)
```
