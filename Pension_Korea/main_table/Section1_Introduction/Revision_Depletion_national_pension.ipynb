```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = [2023, 2030, 2040, 2050, 2060, 2070, 2080, 2088, 2093]
total_people = [2199, 2090, 1843, 1534, 1251, 1086, 942, 901, 861]
dependent_people = [527, 761, 1160, 1467, 1569, 1501, 1347, 1154, 1030]
percent_dependent = [24.0, 36.4, 62.9, 95.6, 125.4, 138.3, 143.1, 128.1, 119.6]

bar_width = 0.35
index = np.arange(len(years))

fig, ax1 = plt.subplots(figsize=(8, 6))


# Bar chart for number of people
bar1 = ax1.bar(index, total_people, bar_width, alpha=0.6, label="National Pension contributor", color='#E95793')
bar2 = ax1.bar(index + bar_width, dependent_people, bar_width, alpha=0.6, label="National Pension recipient", color="#940B92")
ax1.set_ylabel('The numbers (Unit: ten thousands)', fontsize=14)
ax1.set_xlabel('Years', fontsize=14)
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(years, fontsize=13)
ax1.legend(loc='upper left', fontsize=13)
ax1.set_ylim(0, 2800)

# Create a second y-axis for percent dependent
ax2 = ax1.twinx()
ax2.plot(index + bar_width / 2, percent_dependent, color='#610C9F', marker='o', label='Dependency ratio')
ax2.set_ylabel('Percent (%)', fontsize=14)
ax2.set_ylim(0, 165)
ax2.legend(loc='upper right', fontsize=13)

ax1.grid(which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)


plt.title('(a) Projected National Pension participans with dependency ratio', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for first plot
years1 = [2023, 2030, 2040, 2050, 2060, 2070, 2080, 2088, 2093]
total_people = [2199, 2090, 1843, 1534, 1251, 1086, 942, 901, 861]
dependent_people = [527, 761, 1160, 1467, 1569, 1501, 1347, 1154, 1030]
percent_dependent1 = [24.0, 36.4, 62.9, 95.6, 125.4, 138.3, 143.1, 128.1, 119.6]
index1 = np.arange(len(years1))

# Data for second plot
years2 = [2020, 2025, 2030, 2035, 2040, 2045, 2050, 2055]
percent_dependent2 = [740, 883.85, 1027.7, 1023.8, 1019.9, 718.15, 416.4, 0]
index2 = np.arange(len(years2))

fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(16, 6))

# First plot
bar_width = 0.35
bar1 = ax1.bar(index1, total_people, bar_width, alpha=0.6, label="National Pension contributor (A)", color='#E95793')
bar2 = ax1.bar(index1 + bar_width, dependent_people, bar_width, alpha=0.6, label="National Pension recipient (B)", color="#940B92")
ax1.set_ylabel('The numbers (Unit: ten thousands)', fontsize=14)
ax1.set_xlabel('Years', fontsize=14)
ax1.set_xticks(index1 + bar_width / 2)
ax1.set_xticklabels(years1, fontsize=13)
ax1.legend(loc='upper left', fontsize=13)
ax1.set_ylim(0, 2800)
ax2 = ax1.twinx()
ax2.plot(index1 + bar_width / 2, percent_dependent1, color='#610C9F', marker='o', label='Dependency ratio (B/A)')
ax2.set_ylabel('Percent (%)', fontsize=14)
ax2.set_ylim(0, 165)
ax2.legend(loc='upper right', fontsize=13)
ax1.grid(which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax1.set_title('(a) Projected National Pension participants with dependency ratio', fontsize=16)

# Second plot
ax3.plot(index2 + bar_width / 2, percent_dependent2, color='#D80032', marker='o', label='National Pension fund')
ax3.axhline(y=0.0, color='black', linestyle='--')
ax3.set_xlabel('Years', fontsize=14)
ax3.set_xticks(index2 + bar_width / 2)
ax3.set_xticklabels(years2, fontsize=13)
ax3.set_ylabel('Amount (Unit: trillion KRW)', fontsize=14)
ax3.set_ylim(-50, 1100)
ax3.legend(loc='upper right', fontsize=13)
ax3.grid(which='major', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax3.set_title('(b) Projected National Pension fund', fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
fig.savefig("national_pension_dep.eps", format="eps", dpi=1200)
```
